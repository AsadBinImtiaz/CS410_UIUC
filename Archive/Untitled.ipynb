```python
def download_file(filename_with_path):
    if os.path.isfile(filename_with_path):
        printTS(f'File: {filename_with_path} Locally exists')
        return
    else:
        printTS(f'Downloading {filename_with_path} from S3 Bucket: {bucket}')
        with open(dir_home+filename_with_path,'wb') as fin:
            s3 = s3fs.S3FileSystem(anon=False)
            if s3.path.isfile(filename_with_path):
                with s3.open(bucket+filename_with_path,'rb') as fout:
                    fin.write(fout.read())
                    printTS(f'File {filename_with_path} dowloaded from S3 Bucket: {bucket}')
            else:
                printTS(f'File {filename_with_path} not exists on S3 Bucket: {bucket} path {filename_with_path}')
```
